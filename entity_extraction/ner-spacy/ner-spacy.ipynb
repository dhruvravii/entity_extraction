```python
name_train = []
company_train = []
place_train = []
time_train = []
money_train = []


with open("names.txt", "r", encoding = "utf-8") as f:
  new_text = f.read()

start = 1

for i,v in enumerate(new_text):
    if(v == '|'):
        data_point = (new_text[start:i], [(0,i - start - 1 , "PERSON")])
        name_train.append(data_point)
        start = i +1
print(name_train)



with open("company.txt", "r", encoding = "utf-8") as f:
  new_text = f.read()

start = 1

for i,v in enumerate(new_text):
    if(v == '|'):
        data_point = (new_text[start:i], [(0,i - start - 1 , "ORG")])
        company_train.append(data_point)
        start = i +1
print(company_train)



with open("place.txt", "r", encoding = "utf-8") as f:
  new_text = f.read()

start = 1

for i,v in enumerate(new_text):
    if(v == '|'):
        data_point = (new_text[start:i], [(0,i - start - 1 , "GPE")])
        place_train.append(data_point)
        start = i +1
print(place_train)



with open("time.txt", "r", encoding = "utf-8") as f:
  new_text = f.read()

start = 1

for i,v in enumerate(new_text):
    if(v == '|'):
        data_point = (new_text[start:i], [(0,i - start - 1 , "MONTH")])
        time_train.append(data_point)
        start = i +1
print(time_train)



with open("money.txt", "r", encoding = "utf-8") as f:
  new_text = f.read()

start = 1

for i,v in enumerate(new_text):
    if(v == '|'):
        data_point = (new_text[start:i], [(0,i - start - 1 , "MONEY")])
        money_train.append(data_point)
        start = i +1
print(money_train)
        
train_data = name_train + company_train + money_train + place_train + time_train

print("Training data final:")
print(train_data)
print(len(train_data))
        
```

```python
from spacy.tokens import DocBin
from tqdm import tqdm
import spacy

nlp = spacy.blank("en")
db = DocBin()
for text, annotations in tqdm(train_data):
    print(text)
    print(annotations)
    doc = nlp.make_doc(text)
    ents = []
    for start, end, label in annotations:
        print(start)
        print(end)
        print(label)
        span = doc.char_span(start, end, label=label, alignment_mode = "expand")
        print(span)
        if span is None:
            print("Skipping Entity")
        else:
            ents.append(span)
    doc.ents = ents
    print(doc.ents)
    db.add(doc)
db.to_disk("./train.spacy")
```
